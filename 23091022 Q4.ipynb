```python
#Q4. Polynomial Plotting with Annotations:
#Randomly select the coefficients of a 5th degree polynomial. 
#Set the random seed as the last two digits of your roll number. 
#Plot the polynomial for -5 ≤ x ≤ 5. Annotate the plot to identify the maxima and minima of the polynomial
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(22)  # Change 24 to your last two digits of your roll number

# Generate random coefficients for a 5th degree polynomial
coefficients = np.random.rand(6)  # 6 coefficients for a 5th degree polynomial

# Define the polynomial function
def polynomial(x, coeffs):
    return np.polyval(coeffs, x)

# Generate x values
x_values = np.linspace(-5, 5, 400)

# Calculate y values using the polynomial function
y_values = polynomial(x_values, coefficients)

# Plot the polynomial
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Polynomial')

# Find critical points (where the derivative equals zero)
critical_points = np.roots(np.polyder(coefficients))

# Filter real critical points within the range -5 to 5
real_critical_points = [x for x in critical_points if -5 <= x <= 5]

# Calculate the corresponding y values for the critical points
y_critical_points = polynomial(real_critical_points, coefficients)

# Annotate maxima and minima
for x, y in zip(real_critical_points, y_critical_points):
    if np.polyder(np.polyder(coefficients))(x) > 0:
        plt.annotate('Maxima', xy=(x, y), xytext=(-30, 20), textcoords='offset points',
                     arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))
    elif np.polyder(np.polyder(coefficients))(x) < 0:
        plt.annotate('Minima', xy=(x, y), xytext=(-30, 20), textcoords='offset points',
                     arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of a Random 5th Degree Polynomial with Annotations')
plt.legend()
plt.grid(True)
plt.show()
```
